```python
import xml.etree.ElementTree as ET
import multiprocessing
from tqdm import tqdm
import pandas as pd
import numpy as np
import requests
import logging
import pickle
import json
import os
```

```python
def processFulltextDocument(paired_path, port=8086):
    file_path, save_path = paired_path
    url = f'http://127.0.0.1:{port}/api/processFulltextDocument'
    header = {'Accept': 'application/xml'}
    try:
        with open(file_path, 'rb') as f:
            files = {'input': f}
            res = requests.post(url=url, files=files, headers=header, timeout=600)
        res.raise_for_status()
        with open(save_path, 'wb') as f:
            f.write(res.content)
    except Exception as e:
        logging.error(f"Error processing {file_path} on port {port}: {e}")
```

```python
def print_tags(element, limit=2, level=0):
    if level >= limit:
        return
    prefix = ' * ' * level
    print(f"{prefix} {element.tag}")
    for child in element:
        print_tags(child, limit, level + 1)
        
def extract_text_with_tags(element):
    return ''.join(element.itertext()).strip()
```

```python
libgen = []
for idx, row in tqdm(table.iterrows(), total=len(table.index)):
    with open(f'/public/home/bdpstu/xintianle/textbooks/libgen/grobid-output/{idx}.xml', 'r', encoding='utf-8') as f:
        xml = ET.fromstring(f.read())
    # abstract
    abstract = []
    for div in xml.find('{http://www.tei-c.org/ns/1.0}teiHeader').find('{http://www.tei-c.org/ns/1.0}profileDesc').find('{http://www.tei-c.org/ns/1.0}abstract').findall('{http://www.tei-c.org/ns/1.0}div'):
        txt = ''
        for part in div:
            if 'head' in part.tag and part.text.strip():
                txt += '## ' + part.text.strip() + '\n'
            else:
                txt += extract_text_with_tags(part).strip() + '\n'
        abstract.append(txt)
    # section
    sections = []
    for section in xml.find('{http://www.tei-c.org/ns/1.0}text').find('{http://www.tei-c.org/ns/1.0}body').findall('{http://www.tei-c.org/ns/1.0}div'):
        txt = ''
        if len(section) == 1 and 'head' in section[0].tag:
            continue
        for part in section:
            text = extract_text_with_tags(part).replace('■', '').replace('\t', '').strip()
            if 'head' in part.tag and text != '':
                txt += '## ' + text + '\n'
            else:
                txt += text + '\n'
        sections.append(txt)
    libgen.append({
        'id': row['id'],
        'md5': idx,
        'title': row['title'],
        'author': row['author'],
        'publisher': row['publisher'],
        'year': row['year'],
        'isbn': row['isbn'],
        'sections': sections
    })
```

```python
with open('/public/home/bdpstu/xintianle/textbooks/libgen/libgen-v2.json', 'w') as f:
    f.write(json.dumps(libgen))
```
